```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# initial values
lr = 0.0001   # how fast to train, bigger not always better
epochs = 100  # how many cylces to train for
expected = 69 # expected answer
```

```python
# forward function
def f(x, y, z):
    return x**2*y+z
```

```python
# loss function
def loss(a):
    return (expected-a)**2

# loss with respect to x, y, z:
# (expected-(x**2*y+z))**2
```

```python
# loss = 0 when a is equal to expected 
plt.plot([loss(a) for a in range(100)])
```

```python
# derivatives
# loss/x, loss/y, loss/z
def dx(): return 2*x*y*dz()
def dy(): return x**2*dz()
def dz(): return -2*(x**2*-y-z+expected)

# alternative
# h = 0.00001
# def dx(x, y, z): return (f(x+h, y, z)-f(x, y, z))/h
```

```python
# initialise values
x, y, z = 1, 2, 3
ls, ds = [], []

# train
for epoch in range(epochs):
    
    # current result
    res = f(x, y, z)
    
    # current loss
    l = loss(res)
    
    #logging
    ls.append(l)
    ds.append((dx(), dy(), dz()))
    
    # tweak values based on derivatives
    x = x - dx() * lr
    y = y - dy() * lr
    z = z - dz() * lr
    
print("Final Result: ", f(x,y,z))
```

```python
# plot losses
plt.plot(ls)
```

```python
# plot derivatives
plt.plot(ds)
```
