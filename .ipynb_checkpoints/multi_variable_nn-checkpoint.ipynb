```python
# variables
h = 0.000001
lr = 0.0001
expected = 69.0
```

```python
def pretty_print(x, y, z, a, l):
    print('{:>20}  {:>20}  {:>20}  {:>20}  {:>20}'.format(x,y,z,a,l))
```

```python
def loss(ans):
    return (expected-ans)**2
```

```python
def forward(x,y,z):
    ans = x**2*y+z
    return ans, loss(ans)
```

```python
def lp(x, y, z):
    return (expected - (x**2*y+z))**2
```

```python
def derive_x(x,y,z):
    return (lp(x+h, y, z)-lp(x, y, z))/h

def derive_y(x,y,z):
    return (lp(x, y+h, z)-lp(x, y, z))/h

def derive_z(x,y,z):
    return (lp(x, y, z+h)-lp(x, y, z))/h
    
```

```python
x, y, z = 1.1, 2.2, 3.3
ans = x**2*y+z
l = 0.0
losses = []
ds = []
vals = []

print('{:>20}  {:>20}  {:>20}  {:>20}  {:>20}'.format('x', 'y', 'z', 'ans', 'loss'))

for i in range(100):
    a, l = forward(x, y, z)
    
    losses.append(l)

    
    # partial differentiation of l

    
    dx = derive_x(x,y,z) * lr
    dy = derive_y(x,y,z) * lr
    dz = derive_z(x,y,z) * lr
    ds.append((dx,dy,dz))
    
    
#    print("x: ", x, "y: ", y, "z: ", z)
#    pretty_print(x,y,z,a,l)
    vals.append((x,y,z))
    
    x -= dx
    y -= dy
    z -= dz
    
pretty_print(x,y,z,forward(x,y,z)[0],l)
```

```python
plt.plot(losses)
plt.show()
```

```python
lines = plt.plot(vals)
plt.legend(iter(lines), ('x', 'y', 'z'))
plt.show()
```

```python
lines = plt.plot(ds)
plt.legend(iter(lines), ('dx', 'dy', 'dz'))
plt.show()
```
