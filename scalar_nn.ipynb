```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ans = 69
h = 0.00001
```

```python
def f(x):
    return 2*x-3

def loss(got):
    return (ans-got)**2*0.01

def lf(x):
    return (ans-(2*x-3))**2

def dx(x):
    return (lf(x+h)-lf(x))/h

def dx_ana(x):
    #f(x) = (ans - 2x-3)**2
    #f'(x) = 2*f(x)*-2
    return 2*f(x)*-2
```

```python
plt.plot([f(x) for x in range(100)])
```

```python
plt.plot([loss(x) for x in range(100)])
```

```python
loss(69)
```

```python
plt.plot([lf(x) for x in range(100)])
```

```python
plt.plot([dx(x) for x in range(100)])
```

```python
x = 2.0
ds = []
ls = []
for i in range(1000):
    out = f(x)
    l = loss(out)
    ls.append(l)
    
    d = dx_ana(out) * 0.1
    ds.append(d)
    x -= d
print(round(out))
```
